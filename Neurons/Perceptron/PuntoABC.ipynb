Perceptrón

```python
# Convertir letras a un formato de entrada y salida
X = np.array([letter.flatten() for letter in letters.values()])
y = np.eye(len(letters))  # Codificación de salida (1-hot encoding)

# Función de activación (escalón)
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Función para hacer predicciones
def perceptron_predict(xi, weights):
    xi = np.insert(xi, 0, 1)  # Insertar el sesgo
    return step_function(np.dot(weights, xi))

# Algoritmo del Perceptrón
def perceptron_train(X, y, learning_rate=0.1, epochs=30):
    weights = np.random.rand(y.shape[1], X.shape[1] + 1)  # +1 por el sesgo
    errors = []

    for epoch in range(epochs):
        total_error = 0
        for xi, target in zip(X, y):
            xi = np.insert(xi, 0, 1)  # Inserto el sesgo
            output = step_function(np.dot(weights, xi))
            error = target - output
            total_error += sum(abs(error))
            weights += learning_rate * np.outer(error, xi)
        
        errors.append(total_error)
        print(f"Época {epoch + 1}/{epochs}, Error Total: {total_error}")
        
        # Se detiene si no hay error
        if total_error == 0:
            print(f"Entrenamiento completado en la época {epoch + 1}")
            break
    return weights, errors

# Función para introducir ruido en las entradas
def add_noise(data, noise_level=0.2):
    noisy_data = data.copy()
    for i in range(noisy_data.shape[0]):
        for j in range(noisy_data.shape[1]):
            if np.random.rand() < noise_level:
                noisy_data[i, j] = 1 - noisy_data[i, j]  # Invertir el bit
    return noisy_data

# Entrenar el perceptrón
weights, errors = perceptron_train(X, y)
```

Gráfica del Error

```python
# Graficar el error global en cada época
plt.plot(range(1, len(errors) + 1), errors, marker='o')
plt.xlabel('Época')
plt.ylabel('Error Global')
plt.title('Error Global del Perceptrón en cada Época')
plt.grid(True)
plt.show()
```

Predicciones sin Ruido

```python
# Función para dibujar la letra predicha
def draw_letter(letter_matrix):
    for row in letter_matrix:
        print(''.join(['#' if val == 1 else ' ' for val in row]))


# Probar el perceptrón con las letras (sin ruido)
print("Predicciones sin ruido:")
for xi, letter in zip(X, letters.keys()):
    prediction = perceptron_predict(xi, weights)
    pred_letter = list(letters.keys())[np.argmax(prediction)]
    print(f"\nLetra esperada: {letter}")
    print(f"Matriz ingresada:\n{xi.reshape(5, 5)}")
    print(f"Letra predicha:")
    draw_letter(letters[pred_letter])
```

Predicciones con Ruido

```python
# Probar el perceptrón con las letras (con ruido)
noisy_X = add_noise(X, noise_level=0.2)
print("\nPredicciones con ruido:")
for xi, letter in zip(noisy_X, letters.keys()):
    prediction = perceptron_predict(xi, weights)
    pred_letter = list(letters.keys())[np.argmax(prediction)]
    print(f"\nLetra esperada: {letter}")
    print(f"Matriz ingresada con ruido:\n{xi.reshape(5, 5)}")
    print(f"Letra predicha:")
    draw_letter(letters[pred_letter])
```
